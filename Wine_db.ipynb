```python
# Data source 1:
url = 'https://www.winemag.com/region/portugal/'
browser.visit(url)
```

```python
# HTML object
html = browser.html

# Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

# Retrieve all elements that contain wine information
results = soup.find_all('li', class_='review-item')
titles = soup.find_all('h3', class_='title')
ratings = soup.find_all('span', class_='rating')

# Iterate through the page
for title in titles:
    print(title.text)
for rating in ratings:
    print(rating.text)
```

```python
# Create a list
wine_review = []

for result in results:
    
    # Use Beautiful Soup's find() method to navigate and retrieve attributes
    titles = result.find('h3', class_='title')
    ratings = result.find('span', class_='rating')
   
    # Insert information in Dictionary
    wine = {
        'titles':titles.text,
        'ratings':ratings.text
     }
    wine_review.append(wine)
    print(wine_review)
```

```python
# Data source 2:
wine_file = "Resources/wine.csv"
df_wine = pd.read_csv(wine_file)

print("Total records ", df_wine['winery'].count())
df_wine.head()
```

```python
# Only keep required columns and rename them, change US to United States and fill region with a value to allow aggregate function later
columns = ['Unnamed: 0','country','province','region_1','title','variety','winery','points','price']
wine_db = df_wine[columns].copy()

wine_db = wine_db.rename(columns={'Unnamed: 0':'Wine_id',
                                            'country': 'Country',
                                            'province': 'Province',
                                            'region_1':'Region_name',
                                            'title':'Wine',
                                            'variety':'Wine_type',
                                            'winery':'Winery',
                                            'points':'Rating',
                                            'price':'Price',
                                            'description':'Description',
                                            'designation': 'Designation'})

wine_db = wine_db.replace(to_replace='US', value='United States')
wine_db['Region_name'].fillna("No region", inplace = True)
wine_db.head()
```

```python
# Drop any records that have null values in any of their column fields
wine_db = wine_db.dropna(how='any')

print("After dropping null values: ") 
print(wine_db.count())
```

```python
# Group wine_data and calculate average price and rating per unique wine in the database)
wine_data = wine_db.groupby(['Wine_id' ,
                                     'Country', 
                                     'Province', 
                                     'Region_name',
                                     'Wine',
                                     'Wine_type',
                                     'Winery'], as_index=False)['Rating','Price'].mean().copy()

wine_data.reset_index(inplace=True)
print("Total records ", wine_data['Country'].count())
wine_data.head()
```
